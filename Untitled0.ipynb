#4 자율주행
자율주행 로봇회사 "스타쉽  테크놀로지"

실리콘밸리의 자율주행 로봇회사 스타쉽 테크놀로지는 우버이츠(30%)등 기존 음식배달 서비스보다 저렴한 수수료(10%)로 음식과 식료품을 배달해준다.
고객이 앱을 설치하고 배송지를 지도에서 설정하면 음식점, 식료품등 배송 가능한 상호가 나오고 심지어는 머리 사진을 찍어 보내면 어울릴만한 염색 키트를 미용실에서 보내준다. 바퀴 6개 달린 아이스박스 크기 로봇에 소매점이 물건을 넣으면 로봇은 빠른 걸음 속도로 이동하고 고객은 로봇이 오는 과정을 앱으로 확인할 수 있다. 로봇이 도착하면 고객은 앱으로 잠금장치를 열고 꺼내는 방식이다.
로봇배송 상용화에 걸림돌은 규제이다. 자율주행 로봇을 배송에 활용하려면 '자율주행차'로 허가를 받아야한다. 하지만 자율주행차를 합법화한 지역은 제한적이라 로봇 배송 업계는 스타쉽처럼 자율주행차 허가를 받는 대신 주행 속도를 늦춰 인도를 주행할 수 있는 허가를 받는다. 또 다른 로봇배송의 장애물은 사람들의 거부감인데 코로나로 인해 이 같은 장애가 사라지고 로봇 배송은 수요가 증가할 것으로 보인다.
